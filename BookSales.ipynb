### Count fiction and non fiction per year

```python
count_y_g = pd.DataFrame(data_f.groupby(['Year', 'Genre']).Year.count())
count_y_g = count_y_g.rename(columns={"Year" : "number"})
count_y_g = count_y_g.reset_index()
count_y_g
```

```python
count_y_g.loc[count_y_g["number"] == count_y_g["number"].min()]

```

```python
count_y_g.loc[count_y_g["number"] == count_y_g["number"].max()]
```

### Let's see what books are free

```python
data_f.loc[data_f['Price'] == 0]

#we want to see both genres 
```

### Most valued author

```python
# most reviews


df_reviews = data_f.loc[data_f['Reviews'] > 4000]
df_reviews
```

```python
data_f.loc[data_f['Reviews'] == data_f['Reviews'].max()]
```

#### max books, author

```python
count_books = pd.DataFrame(data_f.groupby('Author').Author.count())
count_books = count_books.rename(columns={"Author" : "number_of_books"}).reset_index()
count_books = count_books.sort_values(by="number_of_books", ascending=False)
count_books.head()
```
